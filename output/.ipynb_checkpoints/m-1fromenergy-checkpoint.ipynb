```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

# Good Ones

```python
# N = 8 Z = 8 O16
x = [0, 0.04, 0.08, 0.12, 0.16]
y = [-76.1367, -76.1368, -76.1371, -76.1376, -76.1383]
ret1 = np.polyfit(x, y, 2)
def generate(array):
    x = np.linspace(0, 0.2, 100)
    return x, [array[2] + array[1] * i + array[0] * i** 2 for i in x]

test = generate(ret1)
plt.plot(x, y,'r.', test[0], test[1], 'g-')
res = np.polyfit(x, y, 2)[0] 
print res
#0.0612 from density
```
