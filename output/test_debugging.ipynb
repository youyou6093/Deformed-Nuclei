```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def plotEnergy(name):
    f = open(name)
    energies = []
    for i in f:
        energies.append(float(i))
    energiesStable = np.array(energies[30:])
    energiesStable -= np.mean(energiesStable)
#     plt.plot(range(len(energies)), energies, 'b.')
#     print energies
#     plt.plot(range(len(energiesStable)), energiesStable, 'r.')
    plt.plot(range(len(energiesStable))[::10], energiesStable[::10], 'r.')
    plt.xlabel("Iterations")
    plt.ylabel("Energy Variation / Mev")
```

```python
plotEnergy('totalenergies0.010000.txt')
```
