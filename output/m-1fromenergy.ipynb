# Testing

```python
def plotEnergy(name):
    f = open(name)
    energies = []
    for i in f:
        energies.append(float(i))
    energiesStable = np.array(energies[35:])
    energiesStable -= np.mean(energiesStable)
#     plt.plot(range(len(energies)), energies, 'b.')
#     print energies
    plt.plot(range(len(energiesStable)), energiesStable, 'b.')
```

```python
plotEnergy("energies40.txt")
```
